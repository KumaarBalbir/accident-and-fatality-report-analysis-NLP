# Data Clearning

```python
data['Age Range'].value_counts()
```

```python
def shift(hr):
  if(hr == 24 or hr<8):
    return 3
  if(hr<16):
    return 1
  return 2
```

```python
def sub_shift(hr):
  s = (hr%8)//2
  di = {0:'A', 1:'B', 2:'C', 3:'D'}
  return f'{shift(hr)}{di[s]}'
```

```python
data['Date'] =  pd.to_datetime(data['Date']) + pd.to_timedelta((data['Time'].astype(str)))
data['SubShifts'] = (data['Date'].dt.hour).apply(sub_shift)
data['Shift'] = (data['Date'].dt.hour).apply(shift)
```

```python
def range_fun(x):
  if(x<30):
    return 'age <30'
  if(x<50):
      return '30<= age <50'
  return 'age >= 50'
```

```python
data['Age Range'] = data['Age of the person'].apply(range_fun)
```

```python
data['Accident Group'] = data['Accident Code'] //100
```

```python
data['Month'] = data['Date'].dt.month
```

```python
data['Day'] = data['Date'].dt.day
```

```python
data = data.drop(columns = {'Time','Gender of the person', 'Time Range', 'No. of persons died along'}, axis = 1)
```

## Analysing the deaths

```python
deaths = data[["Date", "State", "District", "Owner Company", "Mine", "Accident Group"]].groupby(["Date", "State", "District", "Accident Group", "Owner Company", "Mine"], as_index=False).size()
```

### deaths vs. state

```python
fig, ax = plt.subplots(figsize=(18, 8))
ax.tick_params(axis='both', which='major', labelsize=12) 
ax.tick_params(axis='both', which='minor', labelsize=12)
deathvstate = deaths[["State", "size", "Accident Group"]].groupby(["State", "Accident Group"], as_index=False).sum()
deathvstate = deathvstate.pivot(columns="Accident Group", values="size", index="State").fillna(0)
deathvstate.plot.bar(stacked=True, ax=ax)
ax.set_ylabel("Deaths", fontsize=14)
ax.set_xlabel("State", fontsize=14)
ax.set_title("Statewise Death Count", fontsize=18)
ax.legend(["Ground Movement",
           "Transportation Machinery (winding)",
           "Transportation Machinery (non-winding)",
           "Machinery other than transportation machinery",
           "Explosives",
           "Electricity",
           "Dust, gas and other combustible material",
           "Falls (other than fall of ground)",
           "Other causes"], 
          bbox_to_anchor=(1,1), 
          fontsize=15)
```
